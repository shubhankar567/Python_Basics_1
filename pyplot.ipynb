```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x,y,linewidth=1)
plt.plot(x,y2, color='k', linewidth=1, linestyle = '-.', marker = '.',markersize = 10, markeredgecolor = 'b')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("chart")
plt.axis([0,10,-1.5,1.5])
plt.legend(["sin x", "cos x"])
plt.show()
```

```python
# Generate random data

data = np.random.randn(1000)

# Plot histogram
plt.hist(data, bins=20, color='skyblue', edgecolor='black')

# Customize plot
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
# Generate random data for three groups

data1 = np.random.normal(0, 1, 100)


# Combine the data into a list


# Create the box plot
plt.boxplot(data1, vert = False, labels=["Group A"])

# Customize the plot
plt.title('Box Plot')
plt.xlabel('Groups')
plt.ylabel('Values')

# Display the plot
plt.show()
```

```python
# Generate random data for three groups
np.random.seed(42)
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(1, 1.5, 100)
data3 = np.random.normal(2, 0.5, 100)

# Combine the data into a list
data = [data1, data2, data3]

# Create the box plot
plt.boxplot(data)

# Customize the plot
plt.title('Box Plot')
plt.xlabel('Groups')
plt.ylabel('Values')

# Display the plot
plt.show()
```

```python

# Generate random data
np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)

# Create the scatter plot
plt.scatter(x, y, color='red', alpha=0.7, edgecolor='black',marker = 'o')

# Customize the plot
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')

# Display the plot
plt.show()
```
